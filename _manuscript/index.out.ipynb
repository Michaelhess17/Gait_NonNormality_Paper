## Abstract

Healthy walking requires flexible, multi-joint coordination that can absorb perturbations across a range of frequencies and directions while maintaining rhythmic locomotion. Classical gait stability measures – Floquet multipliers and Lyapunov exponents – reduce this rich coordination to a single number summarising how fast perturbations grow or decay, discarding information about *which* perturbation directions matter, *how* the system responds at different frequencies, and *how* the two limbs couple to each other. Here we fit a simple linear model (delay-embedded dynamic mode decomposition) to bilateral hip, knee, and ankle kinematics from 299 treadmill walking trials and show that the structure of this model – specifically, how far it departs from a “well-behaved” (normal) operator – provides a multi-dimensional portrait of locomotor coordination that distinguishes able-bodied, high-functioning stroke, and low-functioning stroke gait in ways that conventional metrics cannot. Able-bodied operators are highly ill-conditioned, enabling the model to respond broadly across stride harmonics and maintain symmetric inter-limb coupling regardless of walking speed. High-functioning stroke operators concentrate their response sharply at the stride frequency, achieving strong resonance but sacrificing frequency breadth. Low-functioning stroke operators are nearly well-conditioned, producing weak responses across all frequencies. Both stroke groups show speed-dependent non-normality (absent in able-bodied gait) and increased inter-limb coupling asymmetry concentrated around the non-paretic knee. These signatures reveal that stroke does not merely slow gait or increase variability: it restructures the dynamical coordination strategy in group-specific ways, suggesting that rehabilitation approaches targeting dynamical richness may complement conventional kinematic goals.

## Introduction

Walking looks simple but requires remarkably sophisticated coordination. Each stride involves dozens of muscles acting across multiple joints on both legs, timed to maintain balance, absorb ground reaction forces, and propel the body forward. When this coordination is disrupted by neurological injury such as stroke, the consequences are visible not just as slower walking or asymmetric steps, but as a qualitative change in how the whole system responds to the moment-to-moment perturbations – from uneven ground, shifting attention, or the body’s own variability – that healthy walkers handle without conscious effort.

How should we measure this coordination? The dominant approach treats walking as a periodic orbit and asks a single question: do perturbations shrink or grow? Floquet multipliers answer this per stride cycle, but require strict periodicity and very large datasets – Revzen and Guckenheimer ([2008](#ref-revzen2008)) needed over 10,000 strides from cockroach running data just to obtain stable estimates, orders of magnitude more than the ~15–50 strides in a typical human trial. They also demand a reliable phase estimate, itself a hard problem ([Revzen and Guckenheimer 2012](#ref-revzen2012); [Bruijn et al. 2013](#ref-bruijn2013)). The maximum Lyapunov exponent ($\lambda_S$) relaxes the periodicity assumption and has emerged as the best-validated predictor of fall risk ([Dingwell and Cusumano 2000](#ref-dingwell2000); [Bruijn et al. 2013](#ref-bruijn2013)), but it provides only a scalar: how fast do nearby trajectories pull apart? It says nothing about *which directions* are unstable, *at what frequencies* the system can respond, or *how the two limbs talk to each other*. These are precisely the questions that matter for understanding locomotor coordination after stroke.

Both measures also share a deeper conceptual issue: they assume the system is autonomous – evolving under its own dynamics without external input. Walking is not autonomous. It is continuously driven by descending neural commands, modulated by sensory feedback, and stabilised by reflexes ([Ting and McKay 2007](#ref-ting2007)). The common finding that $\lambda_S$ is positive (local divergence) despite stable walking is a direct consequence: the system is being simultaneously driven and controlled, and Lyapunov analysis of the autonomous reconstruction picks up the driving, not a genuine instability ([Dingwell and Kang 2007](#ref-dingwell2006)).

We take a different approach. Rather than asking a scalar stability question, we fit a simple linear model to the walking data and then examine the *structure of that model* to learn about the underlying coordination. Specifically, we use dynamic mode decomposition (DMD) – a data-driven technique that finds the best-fit linear operator $A$ predicting the next time step from the current one ([Schmid 2010](#ref-schmid2010); [Tu et al. 2014](#ref-tu2014)) – applied to delay-embedded bilateral joint kinematics. Delay embedding stacks successive time points into a higher-dimensional state vector, allowing the linear model to capture temporal dependencies that a single snapshot cannot ([Takens 1981](#ref-takens1981); [Brunton and Kutz 2019](#ref-brunton2017)).

Like Floquet and Lyapunov analysis, DMD fits an autonomous model. But rather than reducing the result to eigenvalues (which determine long-run stability), we examine the model’s *non-normality*: how far the fitted operator departs from having orthogonal eigenvectors. This departure has concrete physical meaning. A “well-behaved” (normal) operator can only respond to perturbations through its individual modes, each decaying independently. A non-normal operator can transiently amplify specific perturbation directions, respond at frequencies where no single mode is resonant, and couple inputs and outputs in directionally specific ways ([Trefethen and Embree 2005](#ref-trefethen2005); [Henrici 1962](#ref-henrici1962)). These are not abstract mathematical properties – they map directly onto the locomotor questions we care about: How flexibly can the system respond to perturbations across frequencies? How do the two limbs influence each other? And does the system’s coordination structure depend on walking speed, or is it robust across operating conditions?

Non-normality has previously been studied in the DMD literature primarily as a *numerical problem*: Wu, Brunton, and Revzen ([2021](#ref-wu2021)) showed that when DMD is applied to oscillatory data, the underlying operator is inherently non-normal, leading to ill-conditioning and noise-amplified errors. We invert this perspective. Rather than treating non-normality as a nuisance, we show that its magnitude and structure – recovered from walking kinematics – are biologically informative signatures of locomotor coordination. Where the fitted linear model must strain hardest to approximate the true, nonlinear, driven system, non-normality is highest. The pattern of that strain tells us about the coordination it is trying to capture.

Post-stroke gait provides a compelling test case. The clinical picture includes slower walking speeds, inter-limb asymmetry, and elevated stride-to-stride variability ([Bruijn et al. 2013](#ref-bruijn2013)), but these descriptions remain at the level of symptoms. A deeper question is whether stroke fundamentally reorganises *how* the locomotor system coordinates itself – not just how fast or how symmetrically it walks, but the structure of its perturbation responses, its frequency content, and its inter-limb coupling. We test this hypothesis using 299 treadmill walking trials from the open dataset of Winner et al. ([2023](#ref-winner2023)), which includes able-bodied controls and high- and low-functioning stroke survivors. Our approach complements their LSTM-based “gait signatures” framework: where that work used black-box neural networks to identify individual-specific dynamics, we use an explicitly analysable linear operator to extract interpretable coordination signatures.

## Methods

### Dataset

We analyse 299 treadmill walking trials drawn from the open dataset of Winner et al. ([2023](#ref-winner2023)). Each trial comprises 1500 time points (15 s at 100 Hz) of six kinematic channels: bilateral hip, knee, and ankle sagittal-plane joint angles. Trials are partitioned by metadata into able-bodied (AB, $n = 180$; speed range 30–220 cm/s), high-functioning stroke (HF, $n = 65$; 40–127 cm/s), and low-functioning stroke (LF, $n = 54$; 22–60 cm/s). All channels are z-scored within each trial prior to analysis, so that operator properties reflect dynamical organisation rather than absolute amplitude differences.

### Delay-Embedded Hankel DMD

We construct a delay-embedded state by stacking $\tau = 10$ lags of the 6-channel observation, yielding a $66$-dimensional augmented state vector at each time step ([Takens 1981](#ref-takens1981)). This embedding lag spans approximately one stride period at typical walking cadences (100 Hz sampling), ensuring that the augmented space captures at least one full cycle of the dominant oscillation. The resulting Hankel matrix $H \in \mathbb{R}^{66 \times (T - \tau)}$ is split into input ($X = H_{:,1:end-1}$) and output ($Y = H_{:,2:end}$) matrices. We compute the truncated SVD $X = U \Sigma V^*$, retaining the $r$ singular values exceeding $10^{-6}$ times the leading singular value (typically retaining \>99% of the variance), and solve for the reduced operator $\tilde{A} = U_r^* Y V_r \Sigma_r^{-1}$ following the standard projected DMD formulation ([Tu et al. 2014](#ref-tu2014); [Schmid 2010](#ref-schmid2010)). To enforce stability, eigenvalues of $\tilde{A}$ with modulus exceeding unity are projected onto the unit circle. All subsequent analyses operate on this reduced operator.

### Dynamical Quantities

**Henrici departure from normality.** We quantify global non-normality as $\eta = \|A^* A - A A^*\|_2 \,/\, \|A\|_2^2$, where $\|\cdot\|_2$ denotes the spectral (operator 2-) norm. A normal operator satisfies $\eta = 0$; larger values indicate stronger non-orthogonality of eigenvectors and correspondingly greater potential for transient amplification ([Henrici 1962](#ref-henrici1962); [Trefethen and Embree 2005](#ref-trefethen2005)).

**Harmonic resolvent gain.** The stride frequency $f_s$ is identified as the dominant peak (above 0.3 Hz) in the resolvent gain $\sigma_{\max}^{-1}(e^{i\theta}I - A)$ swept around the unit circle. Harmonic resolvent gains are then evaluated at $\|(e^{i 2\pi h f_s / f_{\text{samp}}} I - A)^{-1}\|$ for harmonics $h = 1, \ldots, 5$, quantifying how strongly the operator amplifies periodic forcing at each multiple of the stride frequency.

**Spatial resolvent asymmetry.** The full resolvent at the stride frequency is mapped back to the original $66$-dimensional (delay-embedded) joint space via $R = U \,(z I - \tilde{A})^{-1}\, U^*$. We extract sub-blocks corresponding to paretic$\to$non-paretic and non-paretic$\to$paretic coupling (across all delay lags) and compute their operator norms. The asymmetry index is $\max(G_{12}, G_{21}) / \min(G_{12}, G_{21})$, where $G_{12}$ and $G_{21}$ are the respective sub-block norms. Signed log-ratios of pairwise joint gains quantify the directionality of inter-joint coupling.

**Peak Q-factor.** The quality factor at the stride frequency is $Q = f_s / \Delta f$, where $\Delta f$ is the half-power bandwidth (frequency interval over which the resolvent gain exceeds $\text{peak} / \sqrt{2}$). Higher Q indicates narrower frequency tuning.

**Operator condition number.** The condition number $\kappa(A) = \sigma_{\max}(A) / \sigma_{\min}(A)$ quantifies the ill-conditioning of the operator. High $\kappa$ implies highly non-orthogonal eigenvectors and determines the extent of *pseudospectral spreading*: how far the resolvent gain extends away from eigenvalue locations on the unit circle. For a normal operator, $\kappa = 1$ and the resolvent is determined entirely by eigenvalue distances; for a non-normal operator, the pseudospectral $\varepsilon$-contours broaden with $\kappa$, elevating gain at frequencies where no eigenvalue sits nearby ([Trefethen and Embree 2005](#ref-trefethen2005)).

**Eigenvalue–stride distance.** We also report $\min_j |\lambda_j - e^{i 2\pi f_s / f_{\text{samp}}}|$, the distance from the nearest eigenvalue of $\tilde{A}$ to the stride-frequency point on the unit circle. This quantity controls the peak resolvent gain at the fundamental: tighter proximity yields larger absolute gain, independent of non-normality.

All quantities are computed trial-by-trial. Speed-matched comparisons cap AB trials at the maximum LF walking speed (60 cm/s) to control for the confound that AB subjects walk faster on average.

### Statistical Inference

Because each participant contributes multiple trials at different walking speeds, inferential comparisons use a hierarchical bootstrap rather than treating trials as independent replicates. Within each group, subjects are resampled with replacement, and then that subject’s trials are resampled with replacement within the bootstrap draw. This preserves the nested subject/trial structure while accommodating unequal trial counts per subject. For between-group contrasts we bootstrap differences in medians; for speed-dependence we bootstrap Pearson correlations; and for directional quantities expressed as log-ratios we bootstrap the mean log-ratio. We report percentile 95% confidence intervals and two-sided bootstrap p-values computed from the fraction of bootstrap samples on either side of zero. Unless otherwise noted, cross-group inference is reported on the speed-matched subset.

## Discussion

Three operator-level signatures emerge from delay-embedded DMD of 299 gait trials, collectively revealing that stroke reorganises–rather than merely degrades–the transient dynamical structure of locomotion.

**Speed-dependent non-normality.** Able-bodied gait maintains a stable, speed-invariant level of non-normality, indicating that the transient complexity of the locomotor operator is robust across operating points. Both stroke groups show elevated speed-dependence ($r = 0.35$ HF, $r = 0.47$ LF vs. $r = 0.11$ AB), meaning that their capacity for transient amplification is fragile–contingent on walking at particular speeds. As discussed in the Introduction, all three analysis frameworks–Floquet multipliers, Lyapunov exponents, and DMD–fit autonomous models to a system that is continuously driven and controlled ([Dingwell and Kang 2007](#ref-dingwell2006); [Ting and McKay 2007](#ref-ting2007)). The maximum Lyapunov exponent $\lambda_S$ remains the best-validated predictor of fall risk ([Bruijn et al. 2013](#ref-bruijn2013)), but as a scalar divergence rate of an autonomous reconstruction it cannot distinguish intrinsic instability from neural driving, nor can it capture the directional or frequency-dependent structure of transient dynamics. The Henrici departure, by indexing how severely the autonomous linear model must distort its eigenvector geometry to fit the driven, nonlinear data, captures a complementary dimension of locomotor organisation. Where $\lambda_S$ summarises the dominant instability direction, the resolvent structure maps the full landscape of transient amplification across frequencies and spatial channels. The speed-dependence of this structure in stroke gait–absent in able-bodied gait–suggests that the model-reality gap (i.e., how “non-autonomous” or “nonlinear” the true system is relative to the linear surrogate) varies with operating point in impaired but not in healthy locomotion.

**Spectral organisation: eigenvalue proximity vs. pseudospectral spreading.** The harmonic resolvent profiles initially appear paradoxical: HF has the highest Henrici departure yet the steepest normalised harmonic fall-off, while LF has the lowest non-normality yet a normalised profile similar to AB. The resolution lies in recognising that two distinct mechanisms shape the resolvent–*eigenvalue proximity* to the stride frequency (which sets absolute gain at each harmonic) and *pseudospectral spreading* via the operator’s condition number (which determines how broadly gain extends away from eigenvalue locations). HF operators park an eigenvalue very close to the stride fundamental, producing massive h1 gain. But HF’s condition number ($\kappa \approx 1{,}200$) is an order of magnitude below AB’s ($\kappa \approx 18{,}700$), so the pseudospectral halo that would spread this gain to neighbouring harmonics is narrower. The steep normalised fall-off in HF therefore reflects disproportionate inflation of the fundamental, not suppression of higher harmonics–absolute gains at h2–h5 are comparable to or exceed AB’s. Conversely, LF operators are nearly well-conditioned ($\kappa \approx 46$), meaning almost no pseudospectral spreading; their flat normalised profile arises simply because no single eigenvalue dominates the stride frequency. AB combines extreme ill-conditioning with moderate eigenvalue–stride coupling to produce both high absolute gain and broad bandwidth–the pseudospectral effects sustain resolvent gain at frequencies far from any eigenvalue. From the perspective of Wu, Brunton, and Revzen ([2021](#ref-wu2021)), the condition number gradient across groups reflects how differently the Koopman operator’s non-normal structure is distorted by neurological impairment. The non-normality they identify as a challenge for DMD convergence here becomes a graded diagnostic: AB dynamics strain the autonomous linear model broadly across the spectrum; HF dynamics strain it acutely at one frequency; LF dynamics strain it least of all.

**Asymmetric coupling focused on the non-paretic knee.** The directionality of inter-limb coupling provides perhaps the most clinically informative signature. At the coarse inter-limb level, LF shows clear paretic-dominant cross-body drive, with HF trending in the same direction. At finer joint resolution, however, the most robust effect is even more specific: hip-knee coupling becomes concentrated on the non-paretic knee, and both stroke groups show strong directional bias into that knee from the hips. Rather than the intact side simply compensating for the impaired side in a diffuse manner, the operator structure suggests that stroke channels transient coupling through a knee-centered pathway on the non-paretic limb. That picture remains consistent with the clinical observation that the “good” leg works harder during stroke gait, but it localises the asymmetry to a specific operator-level route through which perturbations are absorbed and redistributed.

**Relationship to existing stability measures.** Bruijn et al. ([2013](#ref-bruijn2013)) evaluated gait stability measures against four validity criteria: construct validity, convergent validity, predictive validity, and discriminant validity. Floquet multipliers showed poor construct validity for human gait–not only because strict periodicity is violated, but because they require prohibitively large datasets (\>10,000 strides in Revzen and Guckenheimer ([2008](#ref-revzen2008)) vs. ~15–50 strides in typical human trials), demand a reliable phase estimator ([Revzen and Guckenheimer 2008](#ref-revzen2008), [2012](#ref-revzen2012)), and are acutely noise-sensitive. $\lambda_S$ showed good predictive validity but limited discriminant validity between populations, and as Dingwell and Kang ([2007](#ref-dingwell2006)) demonstrated, the coexistence of positive $\lambda_S$ (local divergence) with orbital stability reflects the driven nature of the system rather than a genuine autonomous instability. The operator-level metrics introduced here show substantial discriminant validity, with the strongest and most consistent separations in AB-vs-stroke contrasts, and capture facets of locomotor organisation–transient richness, frequency bandwidth, directional coupling–that no single scalar can represent. Like DMD itself, these metrics inherit the autonomous-model assumption, but they make the strain of that assumption visible: non-normality is high precisely where the linear autonomous model must work hardest to approximate driven, nonlinear dynamics. Establishing their predictive validity for fall risk and their convergent validity with $\lambda_S$ are natural directions for future work.

**Methodological considerations.** Our approach builds on the DMD framework of Schmid ([2010](#ref-schmid2010)) and Tu et al. ([2014](#ref-tu2014)), with delay embedding following Takens ([1981](#ref-takens1981)). Wu, Brunton, and Revzen ([2021](#ref-wu2021)) showed that DMD applied to limit-cycle data faces challenges from Krylov space closure and non-normality-amplified estimation errors, and characterised performance via a dataset metric (number of initial conditions $\times$ trajectory length). Our dataset of 299 trials with 1500 time points each provides substantial coverage; the stability enforcement (eigenvalue projection inside the unit disk) and SVD truncation (retaining \>99% variance) guard against the most severe numerical pathologies they identify. Nevertheless, several limitations apply. The analysis uses a single treadmill dataset; overground walking may produce different operator structure. The choice of embedding lag ($\tau = 10$) is motivated by stride-period considerations at 100 Hz but has not been exhaustively optimised. The linear operator is a surrogate–it captures the Koopman-linearised dynamics faithfully only within the delay-embedded subspace, and nonlinear effects beyond this approximation are not modelled.

**Clinical implications.** Collectively, these results suggest that post-stroke locomotor impairment is not a scalar degradation but a structured reorganisation of transient dynamics. Restoring walking speed or step symmetry may not restore the non-normal richness that characterises healthy gait. The group-specific signatures identified here–fragile non-normality, narrowed bandwidth in HF, and knee-centered asymmetry that is strongest in LF–point toward different rehabilitation targets for different functional levels. Interventions that broaden the transient repertoire, such as perturbation-based balance training, multi-frequency rhythmic cueing, or assistive devices that modulate inter-limb coupling directionality, may prove more effective than approaches targeting kinematic outcomes alone.

**Outlook.** The present work demonstrates that non-normality in data-driven linear operators is not merely a numerical nuisance but a window into the transient organisation of biological movement. Future extensions–incorporating residual DMD for rigorous spectral convergence guarantees ([Colbrook, Ayton, and Szöke 2023](#ref-colbrook2021)), measure-theoretic Koopman formulations ([Mezić 2005](#ref-mezic2005), [2013](#ref-mezic2013)), or hybrid embeddings that combine delay coordinates with nonlinear observables–promise richer operator-level phenotypes for clinical stratification and closed-loop control.
