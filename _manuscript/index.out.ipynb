## Abstract

Healthy walking requires flexible, multi-joint coordination that can absorb perturbations across a range of frequencies and directions while maintaining rhythmic locomotion. Classical gait stability measures – Floquet multipliers and Lyapunov exponents – reduce this rich coordination to a single number summarising how fast perturbations grow or decay, discarding information about *which* perturbation directions matter, *how* the system responds at different frequencies, and *how* the two limbs couple to each other. Here we fit a simple linear model (delay-embedded dynamic mode decomposition) to bilateral hip, knee, and ankle kinematics from 299 treadmill walking trials and show that the structure of this model – specifically, how far it departs from a “well-behaved” (normal) operator – provides a multi-dimensional portrait of locomotor coordination that distinguishes able-bodied, high-functioning stroke, and low-functioning stroke gait in ways that conventional metrics cannot. Able-bodied operators maintain high, speed-invariant non-normality (Henrici departure $\eta$) and occupy a high-dimensional subspace, supporting broad frequency responsiveness across harmonics of the stride cycle. High-functioning stroke operators show speed-dependent $\eta$ and park an eigenvalue very close to the stride frequency, producing sharply tuned resonance at the fundamental but steep harmonic fall-off. Low-functioning stroke operators show rising $\eta$ with speed but weak absolute responsiveness across all frequencies. Both stroke groups show increased inter-limb coupling asymmetry concentrated around the non-paretic knee. These signatures reveal that stroke does not merely slow gait or increase variability: it restructures the dynamical coordination strategy in group-specific ways, suggesting that rehabilitation approaches targeting dynamical richness may complement conventional kinematic goals.

## Introduction

Walking looks simple but requires remarkably sophisticated coordination. Each stride involves dozens of muscles acting across multiple joints on both legs, timed to maintain balance, absorb ground reaction forces, and propel the body forward. When this coordination is disrupted by neurological injury such as stroke, the consequences are visible not just as slower walking or asymmetric steps, but as a qualitative change in how the whole system responds to the moment-to-moment perturbations – from uneven ground, shifting attention, or the body’s own variability – that healthy walkers handle without conscious effort.

How should we measure this coordination? The dominant approach treats walking as a periodic orbit and asks a single question: do perturbations shrink or grow? Floquet multipliers answer this per stride cycle, but require strict periodicity and very large datasets – Revzen and Guckenheimer ([2008](#ref-revzen2008)) needed over 10,000 strides from cockroach running data just to obtain stable estimates, orders of magnitude more than the ~15–50 strides in a typical human trial. They also demand a reliable phase estimate, itself a hard problem ([Revzen and Guckenheimer 2012](#ref-revzen2012); [Bruijn et al. 2013](#ref-bruijn2013)). The maximum Lyapunov exponent ($\lambda_S$) relaxes the periodicity assumption and has emerged as the best-validated predictor of fall risk ([Dingwell and Cusumano 2000](#ref-dingwell2000); [Bruijn et al. 2013](#ref-bruijn2013)), but it provides only a scalar: how fast do nearby trajectories pull apart? It says nothing about *which directions* are unstable, *at what frequencies* the system can respond, or *how the two limbs talk to each other*. These are precisely the questions that matter for understanding locomotor coordination after stroke.

Both measures also share a deeper conceptual issue: they assume the system is autonomous – evolving under its own dynamics without external input. Walking is not autonomous. It is continuously driven by descending neural commands, modulated by sensory feedback, and stabilised by reflexes ([Ting and McKay 2007](#ref-ting2007)). The common finding that $\lambda_S$ is positive (local divergence) despite stable walking is a direct consequence: the system is being simultaneously driven and controlled, and Lyapunov analysis of the autonomous reconstruction picks up the driving, not a genuine instability ([Dingwell and Kang 2007](#ref-dingwell2006)).

We take a different approach. Rather than asking a scalar stability question, we fit a simple linear model to the walking data and then examine the *structure of that model* to learn about the underlying coordination. Specifically, we use dynamic mode decomposition (DMD) – a data-driven technique that finds the best-fit linear operator $A$ predicting the next time step from the current one ([Schmid 2010](#ref-schmid2010); [Tu et al. 2014](#ref-tu2014)) – applied to delay-embedded bilateral joint kinematics. Delay embedding stacks successive time points into a higher-dimensional state vector, allowing the linear model to capture temporal dependencies that a single snapshot cannot ([Takens 1981](#ref-takens1981); [Brunton and Kutz 2019](#ref-brunton2017)).

Like Floquet and Lyapunov analysis, DMD fits an autonomous model. But rather than reducing the result to eigenvalues (which determine long-run stability), we examine the model’s *non-normality*: how far the fitted operator departs from having orthogonal eigenvectors. This departure has concrete physical meaning. A “well-behaved” (normal) operator can only respond to perturbations through its individual modes, each decaying independently. A non-normal operator can transiently amplify specific perturbation directions, respond at frequencies where no single mode is resonant, and couple inputs and outputs in directionally specific ways ([Trefethen and Embree 2005](#ref-trefethen2005); [Henrici 1962](#ref-henrici1962)). These are not abstract mathematical properties – they map directly onto the locomotor questions we care about: How flexibly can the system respond to perturbations across frequencies? How do the two limbs influence each other? And does the system’s coordination structure depend on walking speed, or is it robust across operating conditions?

Non-normality has previously been studied in the DMD literature primarily as a *numerical problem*: Wu, Brunton, and Revzen ([2021](#ref-wu2021)) showed that when DMD is applied to oscillatory data, the underlying operator is inherently non-normal, leading to ill-conditioning and noise-amplified errors. We invert this perspective. Rather than treating non-normality as a nuisance, we show that its magnitude and structure – recovered from walking kinematics – are biologically informative signatures of locomotor coordination. The more complex the underlying dynamics that the linear operator must encode, the more non-normal eigenvector geometry it must deploy to do so. We term this encoding demand “dynamical strain” ($\eta$, quantified as the Henrici departure from normality): a measure of how hard the linear model must work to represent the nonlinear, multi-timescale coordination underlying each walking trial.

Post-stroke gait provides a compelling test case. The clinical picture includes slower walking speeds, inter-limb asymmetry, and elevated stride-to-stride variability ([Bruijn et al. 2013](#ref-bruijn2013)), but these descriptions remain at the level of symptoms. A deeper question is whether stroke fundamentally reorganises *how* the locomotor system coordinates itself – not just how fast or how symmetrically it walks, but the structure of its perturbation responses, its frequency content, and its inter-limb coupling. We test this hypothesis using 299 treadmill walking trials from the open dataset of Winner et al. ([2023](#ref-winner2023)), which includes able-bodied controls and high- and low-functioning stroke survivors. Our approach complements their LSTM-based “gait signatures” framework: where that work used black-box neural networks to identify individual-specific dynamics, we use an explicitly analysable linear operator to extract interpretable coordination signatures.

## Methods

### Dataset

We analyse 299 treadmill walking trials drawn from the open dataset of Winner et al. ([2023](#ref-winner2023)). Each trial comprises 1500 time points (15 s at 100 Hz) of six kinematic channels: bilateral hip, knee, and ankle sagittal-plane joint angles. Trials are partitioned by metadata into able-bodied (AB, $n = 180$; speed range 30–220 cm/s), high-functioning stroke (HF, $n = 65$; 40–127 cm/s), and low-functioning stroke (LF, $n = 54$; 22–60 cm/s). All channels are z-scored within each trial prior to analysis, so that operator properties reflect dynamical organisation rather than absolute amplitude differences.

### Delay-Embedded Hankel DMD

We construct a delay-embedded state by stacking $\tau = 10$ lags of the 6-channel observation, yielding a $66$-dimensional augmented state vector at each time step ([Takens 1981](#ref-takens1981)). This embedding lag spans approximately one stride period at typical walking cadences (100 Hz sampling), ensuring that the augmented space captures at least one full cycle of the dominant oscillation. The resulting Hankel matrix $H \in \mathbb{R}^{66 \times (T - \tau)}$ is split into input ($X = H_{:,1:end-1}$) and output ($Y = H_{:,2:end}$) matrices. We compute the truncated SVD $X = U \Sigma V^*$, retaining the $r$ singular values exceeding $10^{-6}$ times the leading singular value (typically retaining \>99% of the variance), and solve for the reduced operator $\tilde{A} = U_r^* Y V_r \Sigma_r^{-1}$ following the standard projected DMD formulation ([Tu et al. 2014](#ref-tu2014); [Schmid 2010](#ref-schmid2010)). To enforce stability, eigenvalues of $\tilde{A}$ with modulus exceeding unity are projected onto the unit circle. All subsequent analyses operate on this reduced operator.

### Dynamical Quantities

**Henrici departure from normality.** We quantify global non-normality as $\eta = \|A^* A - A A^*\|_2 \,/\, \|A\|_2^2$, where $\|\cdot\|_2$ denotes the spectral (operator 2-) norm. A normal operator satisfies $\eta = 0$; larger values indicate stronger non-orthogonality of eigenvectors and correspondingly greater potential for transient amplification ([Henrici 1962](#ref-henrici1962); [Trefethen and Embree 2005](#ref-trefethen2005)).

**Harmonic resolvent gain.** The stride frequency $f_s$ is identified as the dominant peak (above 0.3 Hz) in the resolvent gain $\sigma_{\max}^{-1}(e^{i\theta}I - A)$ swept around the unit circle. Harmonic resolvent gains are then evaluated at $\|(e^{i 2\pi h f_s / f_{\text{samp}}} I - A)^{-1}\|$ for harmonics $h = 1, \ldots, 5$, quantifying how strongly the operator amplifies periodic forcing at each multiple of the stride frequency.

**Spatial resolvent asymmetry.** The full resolvent at the stride frequency is mapped back to the original $66$-dimensional (delay-embedded) joint space via $R = U \,(z I - \tilde{A})^{-1}\, U^*$. We extract sub-blocks corresponding to paretic$\to$non-paretic and non-paretic$\to$paretic coupling (across all delay lags) and compute their operator norms. The asymmetry index is $\max(G_{12}, G_{21}) / \min(G_{12}, G_{21})$, where $G_{12}$ and $G_{21}$ are the respective sub-block norms. Signed log-ratios of pairwise joint gains quantify the directionality of inter-joint coupling.

**Peak Q-factor.** The quality factor at the stride frequency is $Q = f_s / \Delta f$, where $\Delta f$ is the half-power bandwidth (frequency interval over which the resolvent gain exceeds $\text{peak} / \sqrt{2}$). Higher Q indicates narrower frequency tuning.

**Eigenvalue–stride distance.** We also report $\min_j |\lambda_j - e^{i 2\pi f_s / f_{\text{samp}}}|$, the distance from the nearest eigenvalue of $\tilde{A}$ to the stride-frequency point on the unit circle. This quantity controls the peak resolvent gain at the fundamental: tighter proximity yields larger absolute gain, independent of non-normality.

All quantities are computed trial-by-trial. Speed-matched comparisons cap AB trials at the maximum HF walking speed (127 cm/s) to control for the confound that AB subjects walk faster on average.

### Statistical Inference

Because each participant contributes multiple trials at different walking speeds, inferential comparisons use a hierarchical bootstrap rather than treating trials as independent replicates. Within each group, subjects are resampled with replacement, and then that subject’s trials are resampled with replacement within the bootstrap draw. This preserves the nested subject/trial structure while accommodating unequal trial counts per subject. For between-group contrasts we bootstrap differences in medians; for speed-dependence we bootstrap Pearson correlations; and for directional quantities expressed as log-ratios we bootstrap the mean log-ratio. We report percentile 95% confidence intervals and two-sided bootstrap p-values computed from the fraction of bootstrap samples on either side of zero. Unless otherwise noted, cross-group inference is reported on the speed-matched subset.

In healthy walking, the two legs influence each other roughly symmetrically – perturbations on one side propagate to the other with similar strength in both directions. Stroke disrupts this balance. We quantify inter-limb coupling asymmetry as the ratio between the stronger and weaker cross-body influence directions (a ratio of 1 means perfect symmetry). This ratio rises monotonically from AB (median 1.26) through HF (1.69) to LF (1.89), indicating progressively one-sided coupling between the legs. Hierarchical bootstrap confirms larger asymmetry in both stroke groups relative to AB (HF–AB $\Delta$median = 0.44, 95% CI \[0.18, 0.87\], $p$ = 0.002; LF–AB $\Delta$ = 0.63, \[0.22, 1.71\], $p$ \< 0.001). Moreover, in LF gait the dominant direction of influence runs from the paretic to the non-paretic side (mean log-ratio = 0.6, 95% CI \[0.28, 0.9\], $p$ \< 0.001), with HF trending similarly but less certain (0.21, \[-0.01, 0.45\], $p$ = 0.062). In other words, the impaired limb’s dynamics propagate forcefully across the body, and the intact side must absorb and accommodate those perturbations.

How narrowly tuned is each group’s coordination to the stride frequency? The Q-factor measures this: higher Q means the system responds strongly at stride frequency but poorly at neighbouring frequencies. HF gait is the most sharply tuned (median Q = 69.6 vs. AB 33.5 and LF 31.3), consistent with a coordination strategy that locks onto the stride rhythm but sacrifices adaptability to perturbations at other frequencies. Hierarchical bootstrap confirms higher Q in HF than AB ($\Delta$median = 36.0, 95% CI \[3.1, 74.7\], $p$ = 0.034) and in LF than AB ($\Delta$ = -2.2, \[-18.5, 40.5\], $p$ = 0.947), while the HF–LF contrast is less certain ($p$ = 0.176). LF gait thus combines high spatial asymmetry with moderately narrow tuning – a coordination pattern that is both directionally biased and somewhat rigid.

To localise where in the kinematic chain inter-limb asymmetry is concentrated, we examine hip-to-knee coupling. In AB gait, the hips drive both knees roughly equally. In both stroke groups, this coupling shifts toward the non-paretic knee: the log-ratio of hip-to-non-paretic-knee vs. hip-to-paretic-knee coupling is near zero in AB but positive in stroke (HF–AB $\Delta$median = 0.24, 95% CI \[0.02, 0.65\], $p$ = 0.023; LF–AB $\Delta$ = 0.81, \[0.13, 1.42\], $p$ = 0.02). Both stroke groups also show strong directional bias *into* the non-paretic knee from the hips (HF: mean = 1.09, 95% CI \[0.82, 1.36\], $p$ \< 0.001; LF: mean = 1.19, \[0.87, 1.48\], $p$ \< 0.001), with LF exceeding AB ($\Delta$median = 0.69, \[0.22, 1.0\], $p$ = 0.018). A broader summary of all pairwise joint asymmetries trends upward in LF relative to AB ($\Delta$median = 0.33, 95% CI \[0.04, 0.48\], $p$ = 0.022), but the clearest finding is that stroke-related coupling asymmetry is concentrated specifically around the non-paretic knee rather than being diffusely spread across all joints. The intact leg does not simply “compensate” in a general sense – the model reveals a specific pathway through which the non-paretic knee absorbs and redistributes perturbations originating from the impaired side.

## Discussion

Three operator-level signatures emerge from delay-embedded DMD of 299 gait trials, collectively revealing that stroke reorganises–rather than merely degrades–the transient dynamical structure of locomotion.

**Speed-dependent dynamical strain distinguishes coordination strategies.** The central finding is that the dynamical strain ($\eta$) the linear operator must deploy to encode walking kinematics is *speed-invariant* in able-bodied gait but *speed-dependent* in both stroke groups. This difference has a clear interpretation: able-bodied coordination is robust – the same dynamical strategy works across a wide range of walking speeds. In stroke, the coordination strategy itself is fragile: faster walking demands progressively more non-normal encoding, suggesting that speed challenges the damaged system in a way it does not challenge the intact system. HF achieves the highest peak $\eta$ values, indicating that the underlying dynamics are the most complex for the linear model to represent; LF shows rising $\eta$ with speed but lower absolute values, consistent with simpler (but still speed-dependent) underlying coordination.

As discussed in the Introduction, all three analysis frameworks–Floquet multipliers, Lyapunov exponents, and DMD–fit autonomous models to a system that is continuously driven and controlled ([Dingwell and Kang 2007](#ref-dingwell2006); [Ting and McKay 2007](#ref-ting2007)). The maximum Lyapunov exponent $\lambda_S$ remains the best-validated predictor of fall risk ([Bruijn et al. 2013](#ref-bruijn2013)), but as a scalar divergence rate of an autonomous reconstruction it cannot distinguish intrinsic instability from neural driving, nor can it capture the directional or frequency-dependent structure of transient dynamics. Dynamical strain, by indexing how much non-normal eigenvector geometry the autonomous linear model must deploy to encode the driven, nonlinear data, captures a complementary dimension of locomotor organisation. Where $\lambda_S$ summarises the dominant instability direction, the resolvent structure maps the full landscape of transient amplification across frequencies and spatial channels.

**Eigenvalue proximity and dimensionality shape spectral organisation.** The harmonic resolvent profiles initially appear paradoxical: HF has the highest dynamical strain yet the steepest normalised harmonic fall-off. The resolution lies in eigenvalue proximity: HF operators park an eigenvalue very close to the stride frequency on the unit circle, producing massive gain at the fundamental. This tight proximity inflates h=1 so strongly that normalised higher harmonics appear suppressed – but absolute gains at h2–h5 are comparable to or exceed AB’s. The steep normalised fall-off in HF reflects concentrated resonance at one frequency, not a loss of multi-scale responsiveness.

Dimensionality completes this picture. AB operators retain the highest SVD rank, meaning transient amplification can be distributed across more directions and frequencies, supporting the broad harmonic bandwidth observed in Figure 3. HF operators concentrate their high dynamical strain into fewer dimensions, yielding the “narrow but intense” spectral profile – sharp stride-frequency tuning rather than broad harmonic support. LF operators, with both lower dimensionality and lower $\eta$, respond weakly across all frequencies. From the perspective of Wu, Brunton, and Revzen ([2021](#ref-wu2021)), the non-normality they identify as a challenge for DMD convergence here becomes a graded diagnostic: AB dynamics demand broad spectral coverage from the autonomous linear model; HF dynamics demand acute coverage at one frequency; LF dynamics demand the least coverage of all.

**Asymmetric coupling focused on the non-paretic knee.** The directionality of inter-limb coupling provides perhaps the most clinically informative signature. At the coarse inter-limb level, LF shows clear paretic-dominant cross-body drive, with HF trending in the same direction. At finer joint resolution, however, the most robust effect is even more specific: hip-knee coupling becomes concentrated on the non-paretic knee, and both stroke groups show strong directional bias into that knee from the hips. Rather than the intact side simply compensating for the impaired side in a diffuse manner, the operator structure suggests that stroke channels transient coupling through a knee-centered pathway on the non-paretic limb. That picture remains consistent with the clinical observation that the “good” leg works harder during stroke gait, but it localises the asymmetry to a specific operator-level route through which perturbations are absorbed and redistributed.

**Relationship to existing stability measures.** Bruijn et al. ([2013](#ref-bruijn2013)) evaluated gait stability measures against four validity criteria: construct validity, convergent validity, predictive validity, and discriminant validity. Floquet multipliers showed poor construct validity for human gait–not only because strict periodicity is violated, but because they require prohibitively large datasets (\>10,000 strides in Revzen and Guckenheimer ([2008](#ref-revzen2008)) vs. ~15–50 strides in typical human trials), demand a reliable phase estimator ([Revzen and Guckenheimer 2008](#ref-revzen2008), [2012](#ref-revzen2012)), and are acutely noise-sensitive. $\lambda_S$ showed good predictive validity but limited discriminant validity between populations, and as Dingwell and Kang ([2007](#ref-dingwell2006)) demonstrated, the coexistence of positive $\lambda_S$ (local divergence) with orbital stability reflects the driven nature of the system rather than a genuine autonomous instability. The operator-level metrics introduced here show substantial discriminant validity, with the strongest and most consistent separations in AB-vs-stroke contrasts, and capture facets of locomotor organisation–transient richness, frequency bandwidth, directional coupling–that no single scalar can represent. Like DMD itself, these metrics inherit the autonomous-model assumption, but they make the demands of that assumption visible: $\eta$ indexes how much non-normal eigenvector geometry the linear model must deploy to encode the underlying dynamics. Establishing predictive validity for fall risk and convergent validity with $\lambda_S$ are natural directions for future work.

**Methodological considerations.** Our approach builds on the DMD framework of Schmid ([2010](#ref-schmid2010)) and Tu et al. ([2014](#ref-tu2014)), with delay embedding following Takens ([1981](#ref-takens1981)). Wu, Brunton, and Revzen ([2021](#ref-wu2021)) showed that DMD applied to limit-cycle data faces challenges from Krylov space closure and non-normality-amplified estimation errors, and characterised performance via a dataset metric (number of initial conditions $\times$ trajectory length). Our dataset of 299 trials with 1500 time points each provides substantial coverage; the stability enforcement (eigenvalue projection inside the unit disk) and SVD truncation (retaining \>99% variance) guard against the most severe numerical pathologies they identify. Nevertheless, several limitations apply. The analysis uses a single treadmill dataset; overground walking may produce different operator structure. The choice of embedding lag ($\tau = 10$) is motivated by stride-period considerations at 100 Hz but has not been exhaustively optimised. The linear operator is a surrogate–it captures the Koopman-linearised dynamics faithfully only within the delay-embedded subspace, and nonlinear effects beyond this approximation are not modelled. Speed-matched comparisons cap AB trials at the maximum HF walking speed (127 cm/s); cross-group comparisons at higher speeds are not possible because AB is the only group that regularly walks above this threshold.

**Clinical implications.** Collectively, these results suggest that post-stroke locomotor impairment is not a scalar degradation but a structured reorganisation of transient dynamics. Restoring walking speed or step symmetry may not restore the dimensional richness and robust dynamical encoding (speed-invariant $\eta$) that characterise healthy gait. The group-specific signatures identified here–speed-fragile encoding concentrated in few dimensions in HF, rising encoding demand with weak absolute responsiveness in LF, and knee-centered coupling asymmetry that is strongest in LF–point toward different rehabilitation targets for different functional levels. Interventions that broaden the transient repertoire, such as perturbation-based balance training, multi-frequency rhythmic cueing, or assistive devices that modulate inter-limb coupling directionality, may prove more effective than approaches targeting kinematic outcomes alone.

**Outlook.** The present work demonstrates that non-normality in data-driven linear operators is not merely a numerical nuisance but a window into the transient organisation of biological movement. A direct next test of the dimensionality hypothesis is to quantify whether rank mediates spectral breadth: regress decay slope and Q-factor on group with and without rank, and bootstrap within-group partial correlations between rank and (i) decay slope, (ii) h2/h1, and (iii) Q. If HF effects attenuate after conditioning on rank, that would support the “concentrated low-dimensional amplification” mechanism. Future extensions–incorporating residual DMD for rigorous spectral convergence guarantees ([Colbrook, Ayton, and Szöke 2023](#ref-colbrook2021)), measure-theoretic Koopman formulations ([Mezić 2005](#ref-mezic2005), [2013](#ref-mezic2013)), or hybrid embeddings that combine delay coordinates with nonlinear observables–promise richer operator-level phenotypes for clinical stratification and closed-loop control.
